```python
# get time v/s confirmed cases(10-500) and Ti data for a particular country

def get_data(country,l=None):
    #ydata= df[df["Country"]==country].values[0,1:]
    ydata=df.loc[country].values
    ydata = np.array(ydata, dtype=float)
    idx=np.where(ydata>=50)[0][0]
    idx2= np.where(ydata>1)[0][0]
    #ydata=ydata[idx:]
    
    #idx1= np.where(ydata>=1000)[0][0]
    #ydata=ydata[:idx1]
    
    if l!=None:
        ydata = ydata[:l]
    max_days=len(ydata)
    xdata= np.arange(max_days)
    return xdata,ydata, idx
```

```python
# SEIIRR Model Parameter Estimation
def seiirr_ivp(x,y,p,beta0,Tr,Tl,N,lock):
    S,E,I,Rd,Ru = y
    sigma = 1/Tl
    beta=beta0
    if x>lock:
        beta = beta0*0.2 + 0.8*beta0*np.exp(-(x-lock)/7)
    Sd = -beta * (I) * S/(N)
    Ed = beta * (I) * S/(N) - sigma * E
    Id = sigma * E - I*(1+p)/Tr
    Rdd = p*I/Tr
    Rud = I/Tr
    return Sd, Ed, Id, Rdd, Rud

def eval_model_seiirr(params,xdata,ydata,population,t0,lock,min_routine=True): # params: parameters to optimize over: p

    p,beta = params
    N = population
    max_days=len(ydata)
    #Ti=1
    sigma= 1./Tl
    gammaR= 1./Tr
    
    vals=initial_fit[initial_fit["Country"]==country].values[0,1:]
    lam, rd0=vals[0],np.exp(vals[3])
    Rd0=rd0#*np.exp(lam*t0)
    #Rd0 = rd0

    I0= lam/(p*gammaR) * Rd0
    E0 = beta*I0/(lam+sigma)
    Ru0 = gammaR/lam *I0
    S0 = N -(beta/lam * I0)
    '''
    E0 = (Rd0 * (Ti + Td) * Tl) / (p * Ti**2)
    Id0 = Rd0 * Td / Ti
    Iu0 = Rd0 * (1/p-1) * (Ti + Td)/(Ti + Tr) * Tr/Tl
    Ru0 = Rd0 * (1/p-1) * (Ti + Td)/(Ti + Tr)
    S0 = N - (Rd0 + E0 + Id0 + Iu0 + Ru0 )
    '''
    #Ti=10
   # beta=(lam+sigma)*(lam+gammaD)*(lam+gammaR)/(sigma*(lam+gammaD-p*(gammaD-gammaR)))
    if min_routine == True:
        sol = integrate.solve_ivp(seiirr_ivp, [t0, max_days+1], [S0, E0, I0, Rd0, Ru0], args=(p,beta,Tr,Tl,N,lock),t_eval=np.arange(t0,max_days),method='RK45')
        #optim_days = min(100, max_days)  # Days to optimise for
        #weights = 1 / np.arange(1, optim_days+1)[::-1]  
        msle_cases = mean_squared_log_error(ydata[t0:], (sol.y[3,:])) 
        return msle_cases
    else:
        max_days=100
        sol = integrate.solve_ivp(seiirr_ivp, [t0, max_days], [S0, E0, I0, Rd0, Ru0], args=(p,beta,Tr,Tl,N,lock),t_eval=np.arange(t0,max_days),method='RK45')
        return sol
    
import itertools
marker = itertools.cycle(( '*', 'o', '*')) 
no_bounds=(1e-5,1)
def fit_country_seiirr(country, population,bounds=no_bounds,lock=40):
    N = population
    l=None
    xdata,ydata,t0=get_data(country,l)
    #print(t0)
    res_const = minimize(eval_model_seiirr,[10,0.5], bounds=bounds,args=(xdata,ydata, population,t0,lock),method='L-BFGS-B') ## numbers are initial guess
    
    msle_cases=eval_model_seiirr(res_const.x,xdata,ydata,population,t0,lock)
    plt.figure(figsize=(10,9))
    p,beta=res_const.x
    sigma= 1./Tl
    gammaR= 1./Tr
    vals=initial_fit[initial_fit["Country"]==country].values[0,1:]
    lam, intercept=vals[0],vals[3]

    lambdas= 1/2*(-(sigma + (1+p)* gammaR) + np.sqrt((sigma - (1+p)* gammaR)**2 + 4 *beta*sigma))
   # lambdas=lambdas[lambdas>0]
    R0 = beta*(Tr)
    R0obs = beta*Tr/(1+p)
    la = country + ": p = %.6f, "%p +r" $\beta_0$ =  %3.2f, " %beta + 'MSLE = %.3f \n '%msle_cases + '$R_0$ = %.2f, '%R0 + '$R_0^o$ = %.2f, '%R0obs + '$\lambda^{est}$=%.3f'%lambdas 
    #print(Ti)
    sol= eval_model_seiirr(res_const.x,xdata,ydata,population,t0,lock,min_routine=False)

    xdata,ydata,t0=get_data(country)
    ini_fit= np.exp(lam*(xdata-t0)+intercept)
    plt.plot(xdata,ydata,marker=next(marker),linestyle='',label=country,markersize=10)
    plt.plot(xdata,ini_fit,'--',label='initial_fit %.3f'%lam)
    
    plt.plot(sol.t,sol.y[3,:],label = la ,linewidth=2)
    plt.plot(sol.t,sol.y[2,:],label='I')
    plt.plot(sol.t,sol.y[0,:],label='S')

    if lock<80:
        beta_t = beta*0.2 + 0.8*beta*np.exp(-(np.arange(lock,100)-lock)/7)
        plt.plot(np.arange(lock,100),beta_t,label=r'$\beta_t$')
        plt.axvline(lock, linestyle='dashed', color='red',label='lockdown')

    plt.axhline(population,linestyle='dashed',label='population')
    prefix=str(lock) + '_'
    if l!=None :
        plt.axvline(l, linestyle='dashed', color='gray')
        prefix=str(l) + '_'

    plt.yscale('log')
    plt.xlabel('Days')
    plt.ylabel('Confirmed cases fit with $R_d$')
    plt.grid()
    plt.tight_layout()
    plt.legend(fontsize=14)
    #prefix='full_'
    print(country, ': Estimated \lambda =',lambdas)
    plt.savefig(odir+country+'_lock_'+prefix+'p_'+ str(p)[:5]+'.png')
    return sol
```
